En este <i>Notebook</i> se aplica el <b><i>Mtodo de Euler</i></b> con el fin de determinar la evolucin de las concentraciones en el tiempo de las diferentes especies que intervienen en la reaccin de hidrlisis de acetato de etilo. Dicho estudio comenzar con la puesta en marcha de un <b>reactor CSTR</b> y, una vez se alcance el rgimen estacionario, el reactor pasar a funcionar <b>en discontinuo</b> (reactor BATCH). 

En primer lugar, es necesario determinar el sistema de ecuaciones diferenciales ordinarias que describe cmo cambian en el tiempo las concentraciones de las especies que intervienen en la hidrlisis debido a su reaccin qumica.

Para la puesta en marcha del reactor CSTR se ha obtenido el siguiente sistema de EDOs:

$$\frac{dC_A}{dt} = \frac{C_{A0}-C_{A}}{} - kC_{A}C_{B}$$

$$\frac{dC_B}{dt} = \frac{C_{B0}-C_{B}}{} - kC_{A}C_{B}$$

$$\frac{dC_C}{dt} = \frac{-C_{C}}{} + kC_{A}C_{B}$$

$$\frac{dC_D}{dt} = \frac{-C_{D}}{} + kC_{A}C_{B}$$

Por otro lado, para el arranque del reactor discontinuo, se tiene el sistema de EDOs que se muestra a continuacin:

$$\frac{dC_A}{dt} = - kC_{A}C_{B}$$

$$\frac{dC_B}{dt} = - kC_{A}C_{B}$$

$$\frac{dC_C}{dt} =  kC_{A}C_{B}$$

$$\frac{dC_D}{dt} =  kC_{A}C_{B}$$

Como ya se ha mencionado, para resolver los sistemas de EDOs se hace uso del <b>Mtodo de Euler</b>, un algoritmo numrico empleado para aproximar soluciones numricas de ecuaciones diferenciales ordinarias de primer orden. Aplicando este, se han obtenido las siguientes ecuaciones, las cuales permiten determinar la concentracin de cada una de las especies para cada incremento de tiempo:

$$C_{A}=C_{A}+\frac{dC_{A}}{dt} \Delta{t}$$

$$C_{B}=C_{B}+\frac{dC_{B}}{dt} \Delta{t}$$

$$C_{C}=C_{C}+\frac{dC_{C}}{dt} \Delta{t}$$

$$C_{D}=C_{D}+\frac{dC_{D}}{dt} \Delta{t}$$

Todas estas ecuaciones se han implementado en el cdigo de la funcin desarrollada. Dicha funcin devuelve la representacin grfica de la evolucin en el tiempo de las concentraciones de las diferentes especies que intervienen en la reaccin de hidrlisis.
