En este <i>Notebook</i> se aplica el <b><i>Método de Euler</i></b> con el fin de determinar la evolución de las concentraciones en el tiempo de las diferentes especies que intervienen en la reacción de hidrólisis de acetato de etilo. Dicho estudio comenzará con la puesta en marcha de un <b>reactor CSTR</b> y, una vez se alcance el régimen estacionario, el reactor pasará a funcionar <b>en discontinuo</b> (reactor BATCH). 

En primer lugar, es necesario determinar el sistema de ecuaciones diferenciales ordinarias que describe cómo cambian en el tiempo las concentraciones de las especies que intervienen en la hidrólisis debido a su reacción química.

Para la puesta en marcha del reactor CSTR se ha obtenido el siguiente sistema de EDOs:

$$\frac{dC_A}{dt} = \frac{C_{A0}-C_{A}}{τ} - k·C_{A}·C_{B}$$

$$\frac{dC_B}{dt} = \frac{C_{B0}-C_{B}}{τ} - k·C_{A}·C_{B}$$

$$\frac{dC_C}{dt} = \frac{-C_{C}}{τ} + k·C_{A}·C_{B}$$

$$\frac{dC_D}{dt} = \frac{-C_{D}}{τ} + k·C_{A}·C_{B}$$

Por otro lado, para el arranque del reactor discontinuo, se tiene el sistema de EDOs que se muestra a continuación:

$$\frac{dC_A}{dt} = - k·C_{A}·C_{B}$$

$$\frac{dC_B}{dt} = - k·C_{A}·C_{B}$$

$$\frac{dC_C}{dt} =  k·C_{A}·C_{B}$$

$$\frac{dC_D}{dt} =  k·C_{A}·C_{B}$$

Como ya se ha mencionado, para resolver los sistemas de EDOs se hace uso del <b>Método de Euler</b>, un algoritmo numérico empleado para aproximar soluciones numéricas de ecuaciones diferenciales ordinarias de primer orden. Aplicando este, se han obtenido las siguientes ecuaciones, las cuales permiten determinar la concentración de cada una de las especies para cada incremento de tiempo:

$$C_{A}=C_{A}+\frac{dC_{A}}{dt}· \Delta{t}$$

$$C_{B}=C_{B}+\frac{dC_{B}}{dt}· \Delta{t}$$

$$C_{C}=C_{C}+\frac{dC_{C}}{dt}· \Delta{t}$$

$$C_{D}=C_{D}+\frac{dC_{D}}{dt}· \Delta{t}$$

Todas estas ecuaciones se han implementado en el código de la función desarrollada. Dicha función devuelve la representación gráfica de la evolución en el tiempo de las concentraciones de las diferentes especies que intervienen en la reacción de hidrólisis.
